**Cleaning**


```python
gender_map = {
    'f':'Female',
    'F':'Female',
    'M':'Male',
    'm':'Male'
}

df['gender'] = df['gender'].replace(gender_map)
df['gender'] = df['gender'].fillna('Unspecified')
print(f"Gender:{df['gender'].unique()}")
```

```python
df['address_state'] = df['address_state'].fillna('N/A')
print(df['address_state'])

print(df['address_state'])
```

```python
df['helix_community'] = df['helix_community'].fillna(df['helix_community'].median())
print(df['helix_community'])
```
