```python
donations = pd.read_csv("donations.csv")
communications = pd.read_csv("communications.csv")

donations['donation_date'] = pd.to_datetime(donations['donation_date'])
communications['contact_date'] = pd.to_datetime(communications['contact_date'])

donations['year'] = donations['donation_date'].dt.year
communications['year'] = communications['contact_date'].dt.year

donation_trend = donations.groupby('year').agg(
    total_donations=('donation_amount', 'sum'),
    number_of_donations=('donation_id', 'count'),
    unique_supporters=('supporter_id', 'nunique')
).reset_index()

donation_trend

plt.figure(figsize=(8,5))
plt.plot(donation_trend['year'], donation_trend['total_donations'], marker='o')

plt.title("Total Donations Over Time")
plt.xlabel("Year")
plt.ylabel("Total Donation Amount")
plt.grid(True)
plt.xticks(donation_trend['year'])

plt.show()


```

```python
plt.figure(figsize=(8,5))
plt.plot(donation_trend['year'], donation_trend['number_of_donations'], marker='o')

plt.title("Number of Donations Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Donations")
plt.grid(True)
plt.xticks(donation_trend['year'])

plt.show()
```

```python

engagement_trend = communications.groupby('year').agg(
    total_contacts=('contact_id', 'count'),
    opened_messages=('opened', 'sum'),
    clicked_messages=('clicked', 'sum')
).reset_index()

engagement_trend['open_rate'] = (
    engagement_trend['opened_messages'] / engagement_trend['total_contacts']
) * 100

engagement_trend['click_rate'] = (
    engagement_trend['clicked_messages'] / engagement_trend['total_contacts']
) * 100

engagement_trend

plt.figure(figsize=(8,5))
plt.plot(engagement_trend['year'], engagement_trend['open_rate'], marker='o', label='Open Rate')
plt.plot(engagement_trend['year'], engagement_trend['click_rate'], marker='o', label='Click Rate')

plt.title("Supporter Engagement Over Time")
plt.xlabel("Year")
plt.ylabel("Rate (%)")
plt.legend()
plt.grid(True)
plt.xticks(engagement_trend['year'])

plt.show()
```

```python
## Link between engagement and donations 

combined = pd.merge(
    donation_trend,
    engagement_trend,
    on='year'
)

combined

plt.figure(figsize=(10,5))

plt.plot(combined['year'],
         combined['total_donations']/1000000,
         marker='o',
         label='Total Donations (Millions)')

plt.plot(combined['year'],
         combined['open_rate'],
         marker='o',
         label='Open Rate (%)')

plt.plot(combined['year'],
         combined['click_rate'],
         marker='o',
         label='Click Rate (%)')

plt.title("Engagement vs Donation Trends")
plt.xlabel("Year")
plt.ylabel("Value")
plt.legend()
plt.grid(True)

plt.xticks(combined['year'])

plt.show()
```

```python
##Average Donation Value 

donation_trend['average_donation'] = (
    donation_trend['total_donations']
    / donation_trend['number_of_donations']
)

plt.figure(figsize=(8,5))

plt.plot(donation_trend['year'],
         donation_trend['average_donation'],
         marker='o')

plt.title("Average Donation Value Over Time")
plt.xlabel("Year")   
plt.ylabel("Average Donation Amount")

plt.grid(True)

plt.xticks(donation_trend['year'])

plt.show()
```

```python
## Donor Retention vs New Donors

donations_old = pd.read_csv("donations_old.csv")
donations_old['donation_date'] = pd.to_datetime(
    donations_old['donation_date']
)
donations_old['year'] = donations_old['donation_date'].dt.year

all_donations = pd.concat(
    [donations_old, donations],
    ignore_index=True
)

first_donation = all_donations.groupby(
    'supporter_id'
)['year'].min().reset_index()

first_donation.columns = [
    'supporter_id',
    'first_year'
]

donations = donations.drop(
    columns=['first_year', 'first_year_x', 'first_year_y', 'donor_type'],
    errors='ignore'
)

donations = pd.merge(
    donations,
    first_donation,
    on='supporter_id',
    how = 'left'
)

donations['donor_type'] = donations.apply(
    lambda x: 'New Donor'
    if x['year'] == x['first_year']
    else 'Returning Donor',
    axis=1
)

donations[['supporter_id',
           'year',
           'first_year',
           'donor_type']].head()

donor_breakdown = donations.groupby(
    ['year', 'donor_type']
).size().unstack(fill_value=0)

donor_breakdown

donor_breakdown.plot(
    kind='bar',
    stacked=True,
    figsize=(8,5)
)

plt.title("New vs Returning Donors Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Donors")

plt.xticks(rotation=0)

plt.show()
```

```python
#2.
com = pd.read_csv("communications.csv")
com['contact_date'] = pd.to_datetime(
    com['contact_date']
)

channel_engagement = com.groupby(
    'contact_channel'
).agg(
    total_contacts=('contact_id', 'count'),
    opened=('opened', 'sum'),
    clicked=('clicked', 'sum')
).reset_index()

channel_engagement['open_rate'] = (
    channel_engagement['opened']
    / channel_engagement['total_contacts']
) * 100

channel_engagement['click_rate'] = (
    channel_engagement['clicked']
    / channel_engagement['total_contacts']
) * 100

channel_engagement



plt.figure(figsize=(8,5))

x = range(len(channel_engagement))

plt.bar(x,
        channel_engagement['open_rate'],
        width=0.4,
        label='Open Rate')

plt.bar([i + 0.4 for i in x],
        channel_engagement['click_rate'],
        width=0.4,
        label='Click Rate')

plt.xticks([i + 0.2 for i in x],
           channel_engagement['contact_channel'])

plt.title("Engagement by Communication Channel")
plt.xlabel("Channel")
plt.ylabel("Rate (%)")

plt.legend()

plt.show()
```

```python
engagement_donor = pd.merge(
    communications,
    donations[['supporter_id', 'donor_type']],
    on='supporter_id',
    how='left'
)

donor_engagement = engagement_donor.groupby(
    'donor_type'
).agg(
    total_contacts=('contact_id', 'count'),
    opened=('opened', 'sum'),
    clicked=('clicked', 'sum')
).reset_index()

donor_engagement['open_rate'] = (
    donor_engagement['opened']
    / donor_engagement['total_contacts']
) * 100

donor_engagement['click_rate'] = (
    donor_engagement['clicked']
    / donor_engagement['total_contacts']
) * 100

donor_engagement.plot(
    x='donor_type',
    y=['open_rate', 'click_rate'],
    kind='bar',
    figsize=(8,5)
)

plt.title("Engagement by Donor Type")
plt.ylabel("Rate (%)")

plt.xticks(rotation=0)

plt.show()
```

```python
# 3


donor_behaviour = donations.groupby(
    'donor_type'
).agg(
    number_of_donations=('donation_id', 'count'),
    total_donation_amount=('donation_amount', 'sum'),
    average_donation=('donation_amount', 'mean')
).reset_index()

donor_behaviour

plt.figure(figsize=(8,5))

plt.bar(
    donor_behaviour['donor_type'],
    donor_behaviour['average_donation']
)

plt.title("Average Donation Amount by Donor Type")
plt.xlabel("Donor Type")
plt.ylabel("Average Donation Amount")

plt.show()
```

```python
merged_data = pd.merge(
    donations,
    com[['supporter_id', 'contact_channel']],
    on='supporter_id',
    how='left'
)

channel_donation_value = merged_data.groupby(
    'contact_channel'
).agg(
    total_donation_amount=('donation_amount', 'sum')
).reset_index()

channel_donation_value

plt.figure(figsize=(7,7))

plt.pie(
    channel_donation_value['total_donation_amount'],
    labels=[
        f"{channel}\n${amount:,.0f}"
        for channel, amount in zip(
            channel_donation_value['contact_channel'],
            channel_donation_value['total_donation_amount']
        )
    ],
    autopct='%1.1f%%'
)

plt.title("Total Donation Amount by Contact Channel")

plt.show()
```

```python
#4 

channel_analysis = pd.merge(
    channel_engagement,
    channel_donation_value,
    on='contact_channel'
)

channel_analysis

fig, ax1 = plt.subplots(figsize=(10,5))

# Bar chart
ax1.bar(
    channel_analysis['contact_channel'],
    channel_analysis['total_donation_amount'] / 1000000,
    color='skyblue'
)

ax1.set_ylabel("Total Donation Amount (Millions)")
ax1.set_xlabel("Communication Channel")

# Line chart
ax2 = ax1.twinx()

ax2.plot(
    channel_analysis['contact_channel'],
    channel_analysis['click_rate'],
    marker='o',
    linestyle='-',
    linewidth=2,
    color='red'
)

ax2.set_ylabel("Click Rate (%)")

plt.title("Engagement vs Donation Outcomes by Channel")

plt.show()
```

```python
donor_year_relationship = donations.groupby(
    ['year', 'donor_type']
).agg(
    average_donation=('donation_amount', 'mean'),
    unique_supporters=('supporter_id', 'nunique')
).reset_index()

summary = donor_year_relationship.pivot(
    index='year',
    columns='donor_type',
    values=['unique_supporters', 'average_donation']
).fillna(0)

years = summary.index.astype(int)
x = range(len(years))
width = 0.35

fig, ax1 = plt.subplots(figsize=(10, 5))

# Bars: number of supporters
bars1 = ax1.bar(
    [i - width/2 for i in x],
    summary['unique_supporters']['New Donor'],
    width=width,
    label='New Donor Supporters',
    color='steelblue',
    alpha=0.85
)

bars2 = ax1.bar(
    [i + width/2 for i in x],
    summary['unique_supporters']['Returning Donor'],
    width=width,
    label='Returning Donor Supporters',
    color='darkorange',
    alpha=0.85
)

ax1.set_ylabel("Number of Supporters")
ax1.set_xlabel("Year")

# Line: average donation amount
ax2 = ax1.twinx()

line1 = ax2.plot(
    x,
    summary['average_donation']['New Donor'],
    marker='o',
    linewidth=2.5,
    linestyle='--',
    color='green',       
    label='New Donor Avg Donation'
)

line2 = ax2.plot(
    x,
    summary['average_donation']['Returning Donor'],
    marker='o',
    linewidth=2.5,
    linestyle='--',
    color='red',      
    label='Returning Donor Avg Donation'
)

ax2.set_ylabel("Average Donation Amount")

plt.xticks(x, years)


handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
fig.legend(
    handles1 + handles2,
    labels1 + labels2,
    loc='upper center',
    bbox_to_anchor=(0.5, -0.08), 
    ncol=2,
    frameon=True
)

plt.title("Supporters and Average Donation by Donor Type Over Time")
plt.tight_layout()
plt.show()
```

```python
engagement_donor = pd.merge(
    com,
    donations[['supporter_id',
               'donor_type',
               'donation_amount',
               'year']],
    on='supporter_id',
    how='left'
)

donor_year_relationship = engagement_donor.groupby(
    ['year', 'donor_type']
).agg(
    click_rate=('clicked', 'mean'),
    average_donation=('donation_amount', 'mean'),
    unique_supporters=('supporter_id', 'nunique')
).reset_index()

# Convert click rate into percentage
donor_year_relationship['click_rate'] = (
    donor_year_relationship['click_rate'] * 100
)

donor_year_relationship


plt.figure(figsize=(8,5))

for donor in donor_year_relationship['donor_type'].unique():
    
    subset = donor_year_relationship[
        donor_year_relationship['donor_type'] == donor
    ]
    
    plt.plot(
        subset['year'],
        subset['click_rate'],
        marker='o',
        linewidth=3,
        label=donor
    )

plt.title("Click Rate Over Time by Donor Type")
plt.xlabel("Year")
plt.ylabel("Click Rate (%)")


plt.xticks(sorted(donor_year_relationship['year'].unique()))

plt.legend()

plt.grid(True)

plt.show()
```
